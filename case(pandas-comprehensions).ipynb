```python
# Task 16: Write a function that gives 1 to those age's under 30 and 0 to those equal to or above 30. 
# Create a variable named age_flag in the titanic dataset using the function you wrote. (use apply and lambda)
```

```python
def age_controller(age):
    if age < 30:
        return 0
    else:
        return 1

df["age_flag"] = df["age"].apply(lambda x: age_controller(x))
df["age_flag"] = df["age"].apply(lambda x: 1 if x < 30 else 0)
```
